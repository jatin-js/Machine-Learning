```python
import numpy as np
import pandas as pd
```

```python
points = []
sem = ['11', '12', '21', '22']
for i in sem:
    points.append(pd.read_excel('points.xlsx', sheet_name=i))
singlepoint = points[0]
for i in np.arange(len(points)-1):
    singlepoint = pd.concat([singlepoint, points[i+1]], ignore_index=True)
    
```

```python
singlepoint
```

```python
sortedsinglepoint = singlepoint.sort_values(by=['RegNo', 'Year', 'Sem'], ignore_index=True)
```

```python
sortedsinglepoint['ptscre'] = sortedsinglepoint['Pts'] * sortedsinglepoint['Credits']
```

```python
sortedsinglepoint
```

```python
sumD = sortedsinglepoint.groupby(['RegNo', 'Year', 'Sem']).sum()
```

```python
sumD
```

```python
sumD['SGPA'] = sumD['ptscre'] / sumD['Credits']
```
