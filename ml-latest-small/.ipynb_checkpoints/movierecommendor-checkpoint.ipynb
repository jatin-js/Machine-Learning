```python
Sigma = np.zeros((df.shape[0], df.shape[1]))
# populate Sigma with n x n diagonal matrix
Sigma[:df.shape[0], :df.shape[0]] = np.diag(s)
# select
n_elements = 4
Sigma = Sigma[:, :n_elements]
v = v[:n_elements, :]
# reconstruct
B = u.dot(Sigma.dot(v))
print(B)
# transform
T = u.dot(Sigma)
print(T)
T = df.dot(v.T)
print(T)
```
