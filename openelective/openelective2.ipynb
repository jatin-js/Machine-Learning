```python
import pandas as pd
import numpy as np
branches = ['BTE', 'CHE', 'CE', 'CSE', 'EEE', 'ECE', 'ME', 'MME']
sheets = []
subjects = ['BT390', 'CH392', 'CE390', 'CS390', 'EE391', 'EC390', 'ME390', 'MM390', 'SM390', 'HS390', 'MA390', 'PH391', 'ME391']
columns = ['S.No', 'BT390', 'CH392', 'CE390', 'CS390', 'EE391', 'EC390', 'ME390', 'MM390', 'SM390', 'HS390', 'MA390', 'PH391', 'ME391']


for branch in branches:
    sheet = pd.read_excel('new_data.xlsx', sheet_name=branch)
    sheets.append(sheet)
sheets[0]
```

```python
for i in np.arange(len(sheets)):
    for j in subjects:
        if j not in sheets[i].columns:
            sheets[i][j] = len(sheets[i].columns)
```

```python
sheets[6]
```

```python
new_sheets = sheets
writer = pd.ExcelWriter('new_data2.xlsx', engine='xlsxwriter')
for i in range(8):

    new_sheets[i].to_excel(writer, sheet_name=branches[i], index=False)

writer.save()
```

```python
comdf = pd.DataFrame()
for df in new_sheets:
    comdf = pd.concat([comdf, df], ignore_index=True)
```
