## Missing Attributes

```python
# numcol = ['grade', male, preschool, expecrBachelors, motherHS, motherBachelors, motherWork, fatherHS, fatherBachelors, fathc]
col = pisa.columns
# numcol = col.drop('raceeth')
mfcol = col.drop(['minutesPerWeekEnglish', 'studentsInEnglish', 'schoolSize', 'readingScore'])
mediannumcol = ['minutesPerWeekEnglish', 'studentsInEnglish']
meannumcol = ['schoolSize']
```

```python
from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(strategy = "mean")
imputer2 = SimpleImputer(strategy = "median")
imputer3 = SimpleImputer(strategy = "most_frequent")
imputer1.fit(pisa[meannumcol])
imputer2.fit(pisa[mediannumcol])
imputer3.fit(pisa[mfcol])
```

```python
import numpy as np
pisa[meannumcol] = imputer1.transform(pisa[meannumcol])
pisa[mediannumcol] = imputer2.transform(pisa[mediannumcol])
pisa[mfcol] = imputer3.transform(pisa[mfcol])
```

```python
pisa.info()
```

## Separating features and labels

```python
pisa_x = pisa.drop('readingScore', axis=1)
pisa_y = pisa['readingScore'].copy()
```

```python
pisa_x
```

Giving numeric value to string feature

```python
pisa['raceeth'].unique()
```

```python
pisa_x['raceeth'] = pisa_x['raceeth'].astype('category')
pisa_x['raceeth'] = pisa_x['raceeth'].cat.reorder_categories(['White', 'Black', 'Hispanic', 'Asian', 'More than one race',
       'American Indian/Alaska Native',
       'Native Hawaiian/Other Pacific Islander'], ordered=True)
pisa_x['raceeth'] = pisa_x['raceeth'].cat.codes
```

```python
pisa_x['raceeth'].unique()
```

```python
pisa_y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # Linear regression model
```

```python
model.fit(pisa_x, pisa_y)
```

```python
from sklearn.metrics import mean_squared_error
pisa_predictions = model.predict(pisa_x)
model.predict([pisa_x.iloc[1]])
```

```python
mse = mean_squared_error(pisa_y, pisa_predictions) # Mean square error
mse
```

```python
rmse = np.sqrt(mse) # Root mean square error
rmse
```

```python
model.score(pisa_x, pisa_y)
```
